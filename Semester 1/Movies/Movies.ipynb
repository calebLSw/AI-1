```python
def get_revenue_budget(df):
    '''
    Uses the movie id to get data from the "Details" endpoint
    Returns a dataframe of revenue and budget data
    '''
    data = {
        "id":[],
        "budget": [],
        "revenue": []
    }

    for i in df["id"]:
        data["id"].append(i)
        url = "https://api.themoviedb.org/3/movie/"+str(i)+"?language=en-US"

        headers = {
            "accept": "application/json",
            "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiI4Nzk4M2ZiZTk3Njk4OGE5ZTIyMzMwZWJmMDMwYTFhZSIsIm5iZiI6MTcyNzk3MTk0OS45MTA1MjcsInN1YiI6IjYzNzQyNGMyNDZlNzVmMDA5MGE3YmRhYyIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.ZsTpMVaVlF8ioN3-WiYEatqWmnGQhCKlILQv4Xi6Eec"
        }

        response = requests.get(url, headers=headers)
        response = response.json()
        data["budget"].append(response["budget"])
        data["revenue"].append(response["revenue"])
        
    return pd.DataFrame(data)
```

```python
def get_best_fit(df, x, y):

    m = 0
    b = 0
    res = 0
    for i in range(len(df)):
        pred_y = (m * df.loc[0][x]) + b
        act_y = df.log[0][y]
        res = res + (pred_y - act_y)**2
    
    return (m, b, res)
```

```python
get_best_fit(df, "budget", "revenue")
```

```python
import pandas as pd

# Define the function
def best_fit_line_equation(data, x, y):
    x = data["budget"]
    y = data["revenue"]
    n = len(x)
    m = (n * (x * y).sum() - x.sum() * y.sum()) / (n * (x ** 2).sum() - x.sum() ** 2)
    b = (y.sum() - m * x.sum()) / n
    return f"y = {m:.2f}x + {b:.2f}"

# Example DataFrame with correct column names
df = pd.DataFrame()

# Check the columns of the DataFrame
print("DataFrame columns:", sns.regplot(data=df, x="budget", y="revenue"))

# Call the function and print the result
equation = best_fit_line_equation(df, 'budget', 'revenue')
print(equation)
```
